```python
# import the required library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns


# load the dataset 
df = pd.read_csv ("Height_Weight_Age_Sex.csv")
 
df.boxplot(by ='sex', column =['BMI'], grid = False) 
```

```python
plt.figure(figsize=[12,12])
ax = sns.boxplot(x="sex", y="height", data=df)
ax = sns.boxplot(x="sex", y="weight", data=df)
ax = sns.swarmplot(x="sex", y="age", data=df, color=".25")


ax.set_title('...')
ax.set_xlabel('#')
ax.set_ylabel('#')

#grid lines
ax.yaxis.grid(True)
```

```python
plt.figure(figsize=[10,8])
plt.style.use('seaborn-deep')

x = df['height']
y = df['weight']

plt.hist([x, y],  label=['height', 'weight'])
plt.legend(loc='upper left')
plt.title('height vs. weight')
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
plt.figure(figsize=[10,8])

n_bins = 5
x = dfxx['height']
y = dfxx['weight']

plt.hist([x, y],  label=['height', 'weight'])
# Male Height
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

x = dfxx['height']
y = dfxx['weight']
w = dfxx['age']
z = dfxx['BMI']

ax0.hist(x, n_bins, density=True, histtype='bar', color='#4B82BA')
ax0.set_title('xx height')

ax1.hist(y, n_bins, density=True, histtype='bar', stacked=True, color='#AA1D83')
ax1.set_title('xx weight')

ax2.hist(w, n_bins, histtype='bar', stacked=True, color='#FFE002')
ax2.set_title('xx age')

ax3.hist(z, n_bins, histtype='bar', stacked=True, color='#57BD32')
ax3.set_title('xx bmi')

fig.tight_layout()
plt.show()
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

x = dfxy['height']
y = dfxy['weight']
w = dfxy['age']
z = dfxy['BMI']

ax0.hist(x, n_bins, density=True, histtype='bar', color='#4B82BA')
ax0.set_title('xy height')

ax1.hist(y, n_bins, density=True, histtype='bar', stacked=True, color='#AA1D83')
ax1.set_title('xy weight')

ax2.hist(w, n_bins, histtype='bar', stacked=True, color='#FFE002')
ax2.set_title('xy age')

ax3.hist(z, n_bins, histtype='bar', stacked=True, color='#57BD32')
ax3.set_title('xy bmi')

fig.tight_layout()
plt.show()
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize = (14, 10))

w = dfxx['height']
x = dfxx['weight']
y = dfxy['height']
z = dfxy['weight']

ax0.plot(w, 'd', color='#FFBC70', label='xx height')
ax0.set_title('xx height')

ax1.plot(x, 'd', color='#FF8E70', label='xx weight')
ax1.set_title('xx weight')

ax2.plot(y, '*', color='#70FFF0', label='xy height')
ax2.set_title('xy height')

ax3.plot(z, '*', color='#50D0F3', label='xy weight')
ax3.set_title('xy weight')
```

```python
plt.figure(figsize=[12,5])

plt.plot(dfxx['height'], 'd', color='#FFBC70', label='xx height')
plt.plot(dfxx['weight'], 'd', color='#FF8E70', label='xx weight')
plt.plot(dfxy['height'], '*', color='#70FFF0', label='xy height')
plt.plot(dfxy['weight'], '*', color='#50D0F3', label='xy weight')

plt.title('height vs. weight')
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
```
