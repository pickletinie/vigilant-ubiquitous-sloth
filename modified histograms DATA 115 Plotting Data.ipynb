```python
import pandas as pd
PPD = pd.read_csv("./plotting_data.csv")
```

```python
plt.figure(figsize=[12,5])
plt.plot(PPD['Closing Price'],color='#4B82BA')
plt.xlabel('Day')
plt.ylabel('Dollar Value (mil)')
plt.title('Closing Price Over Time')
```

```python
plt.figure(figsize=[12,5])
plt.plot(PPD['Opening Price'],color='#AA1D83')
plt.xlabel('Day')
plt.ylabel('Dollar Value (mil)')
plt.title('Opening Price Over Time')
```

```python
plt.figure(figsize=[12,5])
plt.plot(PPD['Opening Price'],color='#AA1D83',label='Opening Price')
plt.plot(PPD['Closing Price'],color='#4B82BA', label='Closing Price')
plt.xlabel('Day')
plt.ylabel('Dollar Value (mil)')
plt.title('Opening vs. Closing Prices')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=[12,5])
plt.plot(PPD['Opening Price'],'*',color='#AA1D83',label='Opening Price')
plt.plot(PPD['Closing Price'],'*',color='#4B82BA', label='Closing Price')
plt.title('Opening vs. Closing Prices')
plt.ylabel('Dollar Value (mil)')
plt.xlabel('Day')
plt.legend()
```

```python
plt.figure(figsize=[10,8])
plt.style.use('seaborn-deep')

x = PPD['Opening Price']
y = PPD['Closing Price']
#bins = np.linspace(0, 45, 90)

plt.hist([x, y],  label=['Opening', 'Closing'])
plt.legend(loc='upper left')
plt.title('Opening vs. Closing Prices')
plt.ylabel('Dollar Value (mil)')
plt.xlabel('Day')
plt.show()
```

```python
plt.figure(figsize=[18,8])

plt.bar([z-.15 for z in range(90)],x,width=.4,label='Opening',color='#AA1D83')
plt.bar([z+.15 for z in range(90)],y,width=.4,label='Closing',color='#4B82BA')

plt.legend(loc='upper left')
plt.title('Opening vs. Closing Prices')
plt.ylabel('Dollar Value (mil)')
plt.xlabel('Day')
plt.show()
```
